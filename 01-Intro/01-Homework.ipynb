## Q2. Computing duration
- Compute the duration variable. It should contain the duration of a ride in minutes.

- What's the standard deviation of the trips duration in January?

1. 32.59
2. 42.59

### Answer

```python
# Compute trip duration and convert to minutes
df_jan['duration'] = (df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime']).dt.total_seconds() / 60

# Check the standard deviation of the duration column
round(df_jan['duration'].std(), 2)
```

## Q4. One-hot encoding 
- Apply one-hot encoding to the pickup and dropoff location IDs. 
- What's the dimensionality of this matrix (number of columns)?

1. 2
2. 155
3. 345
4. 515
5. 715

### Answer

```python
from sklearn.feature_extraction import DictVectorizer

# Select required columns and convert to string
df_jan_clean["PULocationID"] = df_jan_clean["PULocationID"].astype(str)
df_jan_clean["DOLocationID"] = df_jan_clean["DOLocationID"].astype(str)

# Turn into list of dictionaries
dicts = df_jan_clean[["PULocationID", "DOLocationID"]].to_dict(orient="records")

# Start and fit DictVectorizer
dv = DictVectorizer()
X_train = dv.fit_transform(dicts)

# Review output for matrix dimensionality
X_train.shape
```

## Q5. Training a model

- Train a plain linear regression model with default parameters, where duration is the response variable.
- Calculate the RMSE of the model on the training data...What's the RMSE on train?

1. 3.64
2. 7.64
3. 11.64
4. 16.64

### Answer

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
import numpy as np

# State input features and target
X_train = dv.fit_transform(dicts)
y_train = df_jan_clean["duration"].values

# Train linear regression
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict on training set
y_pred = lr.predict(X_train)

# Evaluate using RMSE
rmse = sqrt(mean_squared_error(y_train, y_pred))
round(rmse, 2)
```

## Q6. Evaluating the model
- Now let's apply this model to the validation dataset (February 2023).
- What's the RMSE on validation?

1. 3.81
2. 7.81
3. 11.81
4. 16.81

### Answer

```python
# Compute trip duration and convert to minutes
df_feb['duration'] = (df_feb['tpep_dropoff_datetime'] - df_feb['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filter durations between 1 and 60 minutes
df_feb_clean = df_feb[(df_feb.duration >= 1) & (df_feb.duration <= 60)].copy()

# Select required columns and convert to string
df_feb_clean["PULocationID"] = df_feb_clean["PULocationID"].astype(str)
df_feb_clean["DOLocationID"] = df_feb_clean["DOLocationID"].astype(str)

# Prepare feature dictionaries
dicts = df_feb_clean[["PULocationID", "DOLocationID"]].to_dict(orient="records")

# Transform validation features using the same DictVectorizer (dv) from training
X_val = dv.transform(dicts)

# Target variable
y_val = df_feb_clean["duration"].values

# Predict
y_pred_val = lr.predict(X_val)

# Evaluate using RMSE
rmse_val = sqrt(mean_squared_error(y_val, y_pred_val))
round(rmse_val, 2)
```
